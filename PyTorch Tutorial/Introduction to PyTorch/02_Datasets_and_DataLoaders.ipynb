## [Loading the Fashion-MNIST Data Set](https://pytorch.org/tutorials/beginner/basics/data_tutorial.html) and [Transforms](https://pytorch.org/tutorials/beginner/basics/transforms_tutorial.html)

Using datasets on PyTorch demands two classes:
- Datasets: stores the samples
- DataLoaders: will interate over the Dataset 

When reading Datasets, we can enter a transformation for the features (**transform**) and one of the targets (**target_transform**)

```python
import torch
from torch.utils.data import Dataset
from torchvision import datasets
from torchvision.transforms import ToTensor

import matplotlib.pyplot as pl
```

Importing the FashionMNIST data set (only the training data)

```python
#Mapping of the target values for the Fashion MNIST dataset
labels_map = {
    0: "T-Shirt",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle Boot",
}

training_data = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor(), #tranforms the images (features) from PIL to Tensors
    target_transform=lambda x: labels_map[x] #transforming the numerical target to the name of the items
)
```

Creating a DataLoader that will:
- release 5 (batch) samples at a time when training
- shuffle the dataset after each epoch

```python
from torch.utils.data import DataLoader

batch_size = 5

train_dataloader = DataLoader(training_data, batch_size=5, shuffle=True)
```

Visualizing the images

```python
import matplotlib.pyplot as plt

train_images, train_labels = next(iter(train_dataloader))

for i in range(batch_size):
    plt.imshow(train_images[i].squeeze(), cmap='gray')
    plt.title(train_labels[i])
    plt.show()
```
